# **Solubility of Molecules**
##### Important for chemists and biologists to determine whether a molecule is soluble in water and whether they are good drug candidates #####
###### (inspired by pharmacist sibling!) #####

## Load data (multi variant analysis: y=f(x))


```python
import pandas as pd

data_frame_var = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
data_frame_var
```

## ***Data Preparation for Analysis***

### Data separation as X & Y. Y is target

```python
y = data_frame_var['logS']
y
```

```python
x = data_frame_var.drop('logS', axis =1)
x
```

### Data Splitting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.3, random_state=100)

x_train
```

```python
x_test
```

## **Model Building**

## Linear Regression

#### Training a model

```python
# sciket lear# sklearn has multiple functions as linear regression to prepare dataset and prepare ML models
from sklearn.linear_model import LinearRegression
linear_regression_var = LinearRegression()
# we want to train the empty linearegression model on the following:
linear_regression_var.fit(x_train, y_train)
```

### **Applying the model to make prediction**

```python
y_linReg_train_predition = linear_regression_var.predict(x_train)
y_linReg_test_prediction = linear_regression_var.predict(x_test)

# print(y_linReg_train_predition, y_linReg_test_prediction)
# prediction of training data: y_linReg_train_predition #prediction of testing data: y_linReg_test_prediction
y_linReg_train_predition
```

### **Evaluate Model Performance**

```python
# actual value of y
y_train
```

```python
# predicted value of y
y_linReg_train_predition
```

```python
from sklearn.metrics import mean_squared_error, r2_score

linReg_train_mse = mean_squared_error(y_train, y_linReg_train_predition)
linReg_train_r2 = r2_score(y_train, y_linReg_train_predition)

linReg_test_mse = mean_squared_error(y_test, y_linReg_test_prediction)
linReg_test_r2 = r2_score(y_test, y_linReg_test_prediction)
```

```python
# actual model evaluation:
#training
print('LR MSE (Train): ', linReg_train_mse)
print('LR R2 (Train): ', linReg_train_r2)
#testing
print('LR MSE (Test): ', linReg_test_mse)
print('LR R2 (Test): ', linReg_test_r2)
```

**Implementing Pandas**

```python
linReg_results = pd.DataFrame (['Linear Regression', linReg_train_mse, linReg_train_r2, linReg_test_mse, linReg_test_r2]).transpose()
linReg_results
```

**Renaming Columns:**

```python
linReg_results.columns = ['Methods','Training MSE', 'Training R2', 'Test MSE', 'Test R2']
linReg_results
```

## **Random Forest**


### **Training the Model**
y variable is quantitative, so 'RandomForestRegressor' is used for Regression model. If y was categorical, then the model would have been 'Classification' as RandomForestClassifier

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
# train the model:
rf.fit(x_train, y_train)
```

### **Applying the model to make a prediction**

```python
y_rf_train_predition = rf.predict(x_train)
y_rf_test_prediction = rf.predict(x_test)
```

### **Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_predition)
rf_train_r2 = r2_score(y_train, y_rf_train_predition)

rf_test_mse = mean_squared_error(y_test, y_rf_test_prediction)
rf_test_r2 = r2_score(y_test, y_rf_test_prediction)
```

```python
rf_results = pd.DataFrame (['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Methods','Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

## **Model Comparison**

```python
df_models = pd.concat([linReg_results, rf_results], axis=0)
df_models
```

```python
df_models.reset_index(drop=True)
```
